```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats.mstats import winsorize
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder

pd.set_option('display.max_columns', None)
```

```python
df_test = pd.read_csv('test.csv')
df_train = pd.read_csv('train.csv')

df = pd.concat([df_train, df_test])

df.head()
```

```python
df.info()
```

```python
df_null = pd.DataFrame(df.isnull().sum(), columns=['count']).reset_index()
df_null.loc[df_null['count']>0]
```

```python
# удаление дубликатов
df = df.drop_duplicates()
```

```python
df['Pool'] = df.PoolArea.apply(lambda value: 'yes' if (value > 0) else 'no')
```

```python
house = df[['SaleType', 'SalePrice', 'OverallQual', 'YearBuilt', 'OverallCond', 'GrLivArea', 'GarageCars']].reset_index()
house.info()
```

```python
# исследуем виды распределений случайной величины
sns.pairplot(house, hue='SaleType')
```

```python
plt.figure(figsize=(16,8))
sns.histplot(house['SalePrice'])
plt.title('распределение стоимости')
plt.xlabel('стоимотсть')
plt.ylabel('кол-во')
```

```python
#заполняем пропущенные значения модай, самым часто встречающимся значением
house['SaleType'] = house['SaleType'].fillna(house['SaleType'].mode()[0])
#заполняем пропущенные значения средним значением(применяем при нормальном распределении)
house['SalePrice'] = house['SalePrice'].fillna(house['SalePrice'].mean())
#заполняем пропущенные значения медианным значением(применяем при ассимитричном распределении)
house['GarageCars'] = house['GarageCars'].fillna(house['GarageCars'].median())
```

```python
# определение зависимостей
corr = house.corr()
plt.figure(figsize=(16,8))
sns.set(font_scale=1.4)
sns.heatmap(corr, annot=True,linewidths=0.5,cmap='coolwarm',fmt='.1g')
plt.title('Корреляция')
```

```python
house = df[['GarageType', 'SalePrice']].reset_index()
house['GarageType'] = house['GarageType'].apply(lambda item: {'2Types':0, 'Attchd':1, 'Basment':2, 'BuiltIn':3, 'CarPort':4, 'Detchd':5, 'NA':6}.get(item))
house['GarageType'] = house['GarageType'].fillna(house['GarageType'].mean())
house['SalePrice'] = house['SalePrice'].fillna(house['SalePrice'].mean())
```

```python
x = np.array(house.GarageType.tolist())
y = np.array(house.SalePrice.tolist())
b1 = (np.mean(x * y) - np.mean(x) * np.mean(y)) / (np.mean(x ** 2) - np.mean(x) ** 2)
b0 = np.mean(y) - b1 * np.mean(x)
plt.figure(figsize=(16,8))
plt.scatter(x, y)
plt.title(f"зависмиость цены от сторй. материала = {np.corrcoef(x, y)[1][0]:.2f}")
plt.plot(x, b0 + b1 * x)
plt.xlabel('класс')
plt.ylabel('цена')
plt.show()
```
