##Условие 1: 1 задача
Скачать файл в уроке
Считать данные с помощью pandas
Вывести на экран первые 5 строк
Посмотреть на описание признаков и на их содержание

```python
url = "https://drive.google.com/file/d/1xrx4_QURjMTfpF6OU4nzCPU4q3JdgLDb/view?usp=sharing"
df = pd.read_csv('https://drive.google.com/uc?id=' + url.split('/')[-2])
```

```python
df.head()
```

1.2 Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки дома
Если реновации дома не было, то в новом признаке поставьте 0

```python
df['delta_renovated'] = df.apply(lambda rec: rec['yr_renovated'] - rec['yr_built'] if rec['yr_renovated'] > rec['yr_built'] else 0, axis = 1)
df[['yr_built','yr_renovated','delta_renovated']]
```

1.3 Создайте признаки года продажи, месяца продажи

```python
df['sale_year'] = df['date'].apply(lambda value: datetime.strptime(value, "%Y%m%dT%H%M%S").year)
df['sale_month'] = df['date'].apply(lambda value: datetime.strptime(value, "%Y%m%dT%H%M%S").month)
df[['date','sale_year','sale_month']]
```

1.4 Удалите признаки date, zipcode, lat, long

```python
df.drop(['date', 'zipcode', 'lat', 'long'], axis = 1, inplace = True)
df
```

##Условие 3: 3 задача

3.1 Найдите среднюю стоимость домов в зависимости от количества спален
Отсортируйте от меньшей стоимости к большей

```python
df.groupby('bedrooms')['price'].mean().sort_values()
```

3.2 Найдите минимальную, среднюю и максимальную стоимости домов в зависимости от состояния дома

```python
df.groupby('condition').agg({'price':['min', 'mean', 'max']})
```

3.3 Постройте таблицу с подсчетом количества домов в данных в зависимости от вида на набережную и оценкой вида

```python
df.groupby(['waterfront', 'view'])['id'].count()
```

```python
df.pivot_table(index = 'waterfront', columns = 'view', values = 'id', aggfunc='count')
```
