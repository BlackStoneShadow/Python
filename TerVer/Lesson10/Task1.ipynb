Провести дисперсионный анализ для определения того, есть ли различия среднего
роста среди взрослых футболистов, хоккеистов и штангистов.
Даны значения роста в трех группах случайно выбранных спортсменов:
Футболисты: 173, 175, 180, 178, 177, 185, 183, 182.
Хоккеисты: 177, 179, 180, 188, 177, 172, 171, 184, 180.
Штангисты: 172, 173, 169, 177, 166, 180, 178, 177, 172, 166, 170.

```python
from math import *
from numpy import *
from scipy.stats import *
import matplotlib.pyplot as plt
#анализируемые данные
f = array([173, 175, 180, 178, 177, 185, 183, 182])                 #Футболисты
h = array([177, 179, 180, 188, 177, 172, 171, 184, 180])            #Хоккеисты
w = array([172, 173, 169, 177, 166, 180, 178, 177, 172, 166, 170])  #Штангисты
t = array([*f, *h, *w])
#проверка нормального распределения
n = [shapiro(f)[1], shapiro(h)[1], shapiro(w)[1]]
#проверка равенства дисперсий
b = bartlett(f, h, w)    
print(f"равенство дисперсий Pvalue = {b[1]:.2f}")
print(f"анализ распределения данных Pvalue = [{n[0]:.2f}, {n[1]:.2f}, {n[2]:.2f}]")
if (n[0] > 0.05 and n[1] > 0.05 and n[2] > 0.05 and b[1] > 0.05):  
   
  #Критерий Фишера
  result = f_oneway(f, h, w)    
  #результат
  if(result[1] > 0.05):
    print(f"дисперсионный тест Фишера не выявид статистически значимых различий Fn={result[0]:.2f}, PValue={result[1]:.2f}")
  else:
    print(f"дисперсионный тест Фишера показал наличие статистически значимых различй Fn={result[0]:.2f}, PValue={result[1]:.2f}")  
  
  #Однофакторный дисперсионный анализ
  k = 3
  n = f.shape[0] + h.shape[0] + w.shape[0]
  
  fm = f.mean()
  hm = h.mean()
  wm = w.mean()
  tm = t.mean()

  St = sum((t - tm) ** 2)
  print(f"Сумма квадратных отклонений от общего среднего {St:.2f}")
  Sf = sum((fm - tm) ** 2) * f.shape[0] + sum((hm - tm) ** 2) * h.shape[0] + sum((wm - tm) ** 2) * w.shape[0]
  print(f"Сумма квадратных отклонений средних групповых значений от общего среднего {Sf:.2f}")
  So = sum((f - fm) ** 2) + sum((h - hm) ** 2) + sum((w - wm) ** 2)
  print(f"Остаточная сумма квадратных отклонений {So:.2f}")

  print(f"степень свободы числителя {(k - 1):.2f}")
  print(f"степень свободы знаменателя {(n - k):.2f}")
  Df = Sf / (k - 1)
  print(f"Факторная дисперсия {Df:.2f}")
  Do = So / (n - k)
  print(f"Остаточная дисперсия {Do:.2f}")
  Fn = Df / Do  
  print(f"Критерий Фишера Ft(3.38) < Fn({Fn:.2f}) => принимаем альтернативную гипотизу")
else:      
  result = kruskal(f, h, w)

  if(result[1] > 0.05):
    print(f"отсутствует нормальное распределение данных, непараметрический тест Краскела — Уоллиса не выявид статистически значимых различий PValue={result[1]:.2f}")
  else:
    print(f"отсутствует нормальное распределение данных, непараметрический тест Краскела — Уоллиса показал наличие статистически значимых различй PValue={result[1]:.2f}")    
```
