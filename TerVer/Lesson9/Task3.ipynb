```python
#Даны значения величины заработной платы заемщиков банка (zp) и значения их
#поведенческого кредитного скоринга (ks): zp = [35, 45, 190, 200, 40, 70, 54, 150, 120, 110],
#ks = [401, 574, 874, 919, 459, 739, 653, 902, 746, 832]. 
#Произвести вычисления как в пункте 2, но с вычислением intercept. Учесть, что
#изменение коэффициентов должно производиться на каждом шаге одновременно 
#(то есть изменение одного коэффициента не должно влиять на изменение другого во время одной итерации).
x = array([35, 45, 190, 200, 40, 70, 54, 150, 120, 110])
y = array([401, 574, 874, 919, 459, 739, 653, 902, 746, 832])
alpha = 1e-5
b0,b1 = 0.1,0.1
n = int((x.shape[0] + y.shape[0]) / 2)
#расчет  коэффициента
for i in range(n ** 6):
  z = b0 + b1 * x
  b0 -= alpha * (2 / n) * sum(z - y)
  b1 -= alpha * (2 / n) * sum((z - y) * x)
#график регрессии
plt.scatter(x, y)
plt.title(f"коэффициент b0={b0:.2f}, b1={b1:.2f}")
plt.plot(x, b0 + b1 * x, label="with intercept")
plt.legend(loc='upper center')
plt.show()
```
