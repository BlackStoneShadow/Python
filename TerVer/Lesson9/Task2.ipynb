```python
from math import *
from numpy import *
import matplotlib.pyplot as plt
```

```python
#Даны значения величины заработной платы заемщиков банка (zp) и значения их
#поведенческого кредитного скоринга (ks): zp = [35, 45, 190, 200, 40, 70, 54, 150, 120, 110],
#ks = [401, 574, 874, 919, 459, 739, 653, 902, 746, 832]. Посчитать коэффициент 
#линейной регрессии при заработной плате (zp), используя градиентный спуск (без intercept).
x = array([35, 45, 190, 200, 40, 70, 54, 150, 120, 110])
y = array([401, 574, 874, 919, 459, 739, 653, 902, 746, 832])
alpha = 1e-6
b1 = 0.1
n = int((x.shape[0] + y.shape[0]) / 2)
#расчет  коэффициента
for i in range(n ** 3):
  b1 -= alpha * (2 / n) * sum((b1 * x - y) * x)
#график регрессии
plt.scatter(x, y)
plt.title(f"коэффициент b1={b1:.2f}")
plt.plot(x, b1 * x, label="without intercept")
plt.legend(loc='upper center')
plt.show()
```
