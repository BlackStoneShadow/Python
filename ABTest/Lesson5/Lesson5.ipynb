##Задача 1
Вы провели эксперимент c упрощением формы заказа в магазине Утконос и получили результаты по метрике конверсий в покупку. Выберите метод оценки и оцените есть ли стат.значимые различия между конверсиями в двух группах при alpha = 5%. Дайте краткие рекомендации команде .Результаты: 1)Число юзеров в группах , которые заходили на сайт в период эксперимента: n1 = 15550 и n2 = 15550 . 2) Число юзеров в группах , которые совершили хотя бы одну покупку за период эксперимента: n1 = 164 и n2 = 228 3) Конверсии : conv1 = 1.05% conv2 = 1.47% .

```python
#используем Z критерий при проведении сравнения долей
p1, p2 = 1.05 / 100, 1.47 / 100
n1, n2 = 15_500, 15_500

p0 = 1.96 * (p1 * (1 - p1) / n1 + p2 * (1 - p2) / n2) ** 0.5

cf = [p2 - p1 - p0, p2 - p1 + p0]

if (cf[0] < 0 and cf[1] > 0) :
  print(f"Нет статистически занчимых отличий, доверительный интервал: {cf}")
else:
  print(f"Наблюдаются статистически занчимых различия, доверительный интервал: {cf}")
```

Новая форма заказа существенно увеличила коэффициент конверсии варианта B (1,47%) он оказался на 39,02% выше, чем коэффициент конверсии варианта A (1,05%). Рекомендуется перенести упрощенную форму заказа в рабочую среду.
